**TENSORFLOW**

TensorFlow, Google tarafından geliştirilen açık kaynaklı bir makine öğrenimi ve derin öğrenme (deep learning) çerçevesidir. TensorFlow, büyük ölçekli sayısal hesaplamaları ve veri işlemeyi hızlandırmak için kullanılan çok amaçlı bir kütüphanedir. Özellikle yapay sinir ağları oluşturmak, eğitmek ve kullanmak için yaygın olarak kullanılır.
